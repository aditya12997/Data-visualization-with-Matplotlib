## Loading Necessary Libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Extracting the dataset

```python
gas_prices = pd.read_csv('gas_prices.csv')

gas_prices.head(5)
```

## Checking Null Values

```python
#Checking if we have NAN of null values in the dataset
gas_prices.isna().count()
```

## Creating a plot to visualize the prices of each for country for the range of given years.

```python
plt.figure(figsize=(10,5))

# Looping through all countries and plotting the plot for better visualizations.
for countries in gas_prices:
    if countries != 'Year':
        plt.plot(gas_prices['Year'], gas_prices[countries], marker = 'x', label = countries)

plt.title('Gas prices in USD', fontsize = 20)
plt.xlabel("Year", fontsize =15)
plt.ylabel("Prices in $", fontsize =15)

plt.xticks(gas_prices['Year'][::3])

# moving the legend outside the graph for better visualization
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize = 12)


plt.show()
```

## Plots only for selected countries

```python
plt.figure(figsize=(8,5))

plt.plot(gas_prices['Year'], gas_prices['USA'],'b.-', label = "USA")
plt.plot(gas_prices['Year'], gas_prices['Canada'],'r.-' ,label = "Canada")
plt.plot(gas_prices['Year'], gas_prices['South Korea'],'g.-', label = "South Korea")
plt.plot(gas_prices['Year'], gas_prices['Australia'], 'y.-', label = "Australia")

plt.title('Gas prices in USD', fontdict = {"fontweight": "bold", "fontsize":22 })
plt.xlabel("Year", fontsize =15)
plt.ylabel("Prices in $", fontsize =15)

plt.xticks(gas_prices['Year'][::3])
plt.legend()


plt.show()
```

### Using the Fifa data to analyse various insights

```python
#using Fifa practice dataset
fifa = pd.read_csv("Fifa.csv")
```

```python
fifa = fifa.rename(columns={"ID":"id"})
fifa.head()
```

### Using Histogram to create a visuals for our dataset

```python
#Creating bin ranges for our histogram

bins= [0,20,40,60,80,100]

#plotting the data
plt.hist(fifa['Overall'], bins=bins, color = '#34ebba')

#labelling the data
plt.xticks(bins)
plt.xlabel("Skill Level", fontsize= 15)
plt.ylabel("Number of players", fontsize= 15)
plt.title("Distribution of Player Skills in FIFA 2018", fontdict = {"fontweight": "bold", "fontsize":18 })


plt.show()
```

### Pie charts to analyse the players with different skillsets

```python
plt.figure(1, figsize=(6,6))

left_foot_plyrs = fifa.loc[fifa["Preferred Foot"] == 'Left'].count()[0]
right_foot_plyrs = fifa.loc[fifa["Preferred Foot"] == 'Right'].count()[0]

total = left_foot_plyrs + right_foot_plyrs
left_foot = 100 * (left_foot_plyrs/total)


right_foot = 100 * (right_foot_plyrs/total)


football_plyrs = fifa.loc[fifa["id"]].count()


labels = ['Left_foot', 'Right_foot']

#AUTOPCT helps us get the percantage of the values we want in our pie chart.
plt.pie([left_foot_plyrs, right_foot_plyrs], labels = labels, autopct='%.2f%%')
plt.title("Foot prefernces in our FIFA Players", fontdict = {"fontweight": "bold", "fontsize":15 })

plt.show()
```

```python
weight = fifa['Weight']
weight
```

```python
plt.figure(1, figsize=(6,6))

#Stripped off the lbs from the string given below.
fifa['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in weight]

plt.style.use('ggplot')

#Segregating players according to their weight
light_weight_plyrs = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium_plyrs = fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)].count()[0]
medium_weight_plyrs = fifa.loc[(fifa['Weight'] >= 150) & (fifa['Weight'] < 175)].count()[0]
medium_heavy_plyrs = fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)].count()[0]
heavy_plyrs = fifa.loc[(fifa['Weight'] > 200)].count()[0]


total_weights = [light_medium_plyrs, light_weight_plyrs, medium_weight_plyrs, medium_heavy_plyrs, heavy_plyrs]

labels = ['Under 25', '125-150','150-175','175-200',"Over 200"]

explode = (.4,.1,0,0, .4)


plt.title("Weight Distribution of FIFA PLayers (in lbs)", fontdict = {"fontweight": "bold", "fontsize":15 })
#Plotting the pie chart
plt.pie(total_weights, labels=labels, autopct='%.2f%%', explode= explode)

plt.show()
```

```python
plt.figure(1, figsize=(10,6))

#Checking the ranking of the overall variable where it signifies the ranking of Good/Bad players at each club

barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
arsenal = fifa.loc[fifa['Club'] == 'Arsenal']['Overall']
man_utd = fifa.loc[fifa['Club'] == 'Manchester United']['Overall']

labels = ['FC Barcelona', 'Real Madrid','Arsenal', 'Manchester United']

plt.title("Club Overall Players Ranking", fontdict = {"fontweight": "bold", "fontsize":18 })
plt.boxplot([barcelona, real_madrid, arsenal, man_utd], labels = labels)
plt.xticks(fontsize=15)
plt.ylabel("FIFA Overall Rating", fontsize = 14)

plt.show()
```
